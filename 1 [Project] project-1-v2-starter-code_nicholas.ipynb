## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
#playerID > player_pokemon (maybe more than 1) > pokemon_id > defense or attack
#players[2]["player_pokemon"][11]["defense"]

def cal_power(players, pokedex, player_id):
    total_power = 0
    for a in players[player_id]["player_pokemon"]:
        each_poke_power = players[player_id]["player_pokemon"][a]["attack"] + players[player_id]["player_pokemon"][a]["defense"]
        total_power += each_poke_power
    print(players[player_id]["player_name"],"'s power is ",total_power)
    return(total_power)

# players[player_id]["player_name"],
cal_power(players, pokedex, 1)
````

### 8.1

While you were putting together the prototype code, your colleagues were preparing a dataset of Pokemon and their attributes. (This was a rush job, so they may have picked some crazy values for some...)

The code below loads information from a comma separated value (csv) file. You need to parse this string into a more useable format. The format of the string is:

- Rows are separated by newline characters: \n
- Columns are separated by commas: ,
- All cells in the csv are double quoted. Ex: "PokedexNumber" is the first cell of the first row.


Using for-loops, create a list of lists where each list within the overall list is a row of the csv/matrix, and each element in that list is a cell in that row. Additional criteria:

1. Quotes are removed from each cell item.
2. Numeric column values are converted to floats.
3. There are some cells that are empty and have no information. For these cells put a -1 value in place.

Your end result is effectively a matrix. Each list in the outer list is a row, and the *j*th elements of list together form the *j*th column, which represents a data attribute. The first three lists in your pokedex list should look like this:

    ['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']

````python
# Code to read in pokedex info
raw_pd = ''
pd_data = []
pokedex_file = 'pokedex_basic.csv'
with open(pokedex_file, 'r') as f:
    raw_pd = f.read()
raw_pd_splitline = raw_pd.splitlines()

# seperating by commas
for i,row in enumerate(raw_pd_splitline):
    pd_data.append(raw_pd_splitline[i].split(","))

# removing " and changing from string to float AND changing values if "" to -1

data_clean = []

for ii,row in enumerate(pd_data):
    data_temp = []
    for iii,col in enumerate(row):
        clean_string = col.replace('"',"")
        try:
            if clean_string == "":
                clean_string = -1
            clean_string = float(clean_string)
        except:
            True == True
        data_temp.append(clean_string)
    data_clean.append(data_temp)
data_clean
````

````python
# we have 721 pokemon in the full pokedex
# I assume that everytime we go to a gym, we will see a pokemon and it is a mutually exclusive event (ie, you wont see 2 pokemon and if you see 1 pokemon, you wont be able to see any others)
# assuming probability of seeing a rare pokemon is ~1/10 the probability of seeing a normal pokemon
# hence if we have 110 rare pokemon, the chance of seeing any of these pokemon is 110/7210 while the chance of seeing any of the common ones is 7100(the remaining number)/7210 
# shall use the below probability formula and create a list of the probability of seeing each of the pokemon

rare_list = ["Qwilfish","Bellossom","AmpharosMega Ampharos","Mareep","Igglybuff","Pichu","Ariados","Hoothoot","Volcanion","Sliggoo"]
denominator = len(full_pokedex)*10
p_rare = 1/denominator
print("Probability of seeing a particular rare pokemon is",p_rare)
print(p_rare)
total_p_rare = p_rare*len(rare_list)
p_common = (1 - total_p_rare)/(len(full_pokedex)-len(rare_list))
print("Probability of seeing a particular common pokemon is",p_common)

#check
print(p_common*(len(full_pokedex)-len(rare_list))+p_rare*len(rare_list))

temp_list = [[k,v["name"]] for k, v in full_pokedex.items()]
rarity_list = []
for each_pokemon in temp_list:
    if each_pokemon[1] in rare_list:
        rarity_list.append([each_pokemon[1],"rare",p_rare])
    else:
        rarity_list.append([each_pokemon[1],"not rare",p_common])
rarity_list

#check if the total probability = 1
probability_count = 0
for each_pokemon in rarity_list:
    probability_count += each_pokemon[2]
round(probability_count,10) == 1
    
````
